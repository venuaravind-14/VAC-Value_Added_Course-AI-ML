```python
import pandas as pd
import numpy as np
df = pd.read_csv("amazonreviews.tsv",sep='\t')
df.head()
```

```python
df['label'].value_counts()
```

```python
df.dropna(inplace=True)
```

```python
blanks = []
for i,lb,rv in df.itertuples():
    if(type(rv)==str):
        if(rv.isspace()):
            blanks.append(i)

df.drop(blanks,inplace=True)
df['label'].value_counts()
```

```python
sid.polarity_scores(df.loc[0]['review'])
df.loc[0]['label']
```

```python
df['scores'] = df['review'].apply(lambda review:sid.polarity_scores(review))

df.head()
```

```python
df['compound'] = df['scores'].apply(lambda scoredict:scoredict['compound'])
df.head()
```

```python
sid.polarity_scores(df.loc[0]['review'])
df.loc[0]['scores']
```

```python
df['compscore'] = df['compound'].apply(lambda c:'pos'if c>=0 else 'neg')
df.head()
```

```python
from sklearn.metrics import *

print(df['label'].value_counts())
accuracy_score(df['label'],df['compscore'])
```

```python
print(confusion_matrix(df['label'],df['compscore']))
```

```python
print(classification_report(df['label'],df['compscore']))
```
