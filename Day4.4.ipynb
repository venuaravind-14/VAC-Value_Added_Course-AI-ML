```python
face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/Assets/haarcascade_frontalface_default.xml')

if face_cascade.empty():
    print("Error: Could not load cascade classifier XML file. Please ensure 'haarcascade_frontalface_default.xml' is in the correct path and not corrupted.")
else:
    def detect_face(img):
      face_img=img.copy()
      face_rects=face_cascade.detectMultiScale(face_img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
      for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
      return face_img
    result=detect_face(denis)
    plt.imshow(result,cmap='grey')
    plt.show()
```
